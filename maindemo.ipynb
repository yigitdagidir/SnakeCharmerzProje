```python
import random as rn

#Müşteri

def olasilik():
    return rn.randint(1,100)

class Dukkan():
    def __init__(self,isim, yas,bakiye):
        self.isim = isim
        self.yas = yas
        self.bakiye = bakiye
        
    def urun_sat(self):
        self.urun_fiyat_toplam = 0
        self.alinan_adet = rn.randint(1,10)

        for i in range(self.alinan_adet):
            self.urun_fiyati = rn.randint(1,100)
            self.urun_fiyat_toplam += self.urun_fiyati
        
        self.bakiye += self.urun_fiyat_toplam

        self.bahsis_sans=50  # yükseltilirse olma olasılığı artar.
        self.veresiye_sans=50

        if olasilik()<self.bahsis_sans:
            self.bahsis_al(20)
        elif olasilik()<self.veresiye_sans:
            self.veresiye_ver(20)


    def bahsis_al(self, max_miktar=int):
        self.bakiye += rn.randint(1,max_miktar)
        
    def veresiye_ver(self,max_miktar=int):
        self.bakiye -= rn.randint(1,max_miktar)

oyuncu=Dukkan("Ahmet",20,500)

def main():
    
    print(""" Oyuna Hoşgeldiniz.

            Konum    Başlangıç Bakiye     Veresiye Verme      Tip Alma

    1 -     Beşiktaş     1000                  %40               %60
    2 -     Ümraniye     3000                  %60               %40
    3 -     Fatih        2000                  %50               %50 

    Sayıyı girerek tercihinizi yapınız.""")

    
    def dukkan_tercih():
        while True:
            secim=input("Tercih ettiğiniz seçeneğin numarasını giriniz:")
            if secim=="1":
                oyuncu.bahsis_sans=60
                oyuncu.veresiye_sans=40
                break
            elif secim=="2":
                oyuncu.bahsis_sans=40
                oyuncu.veresiye_sans=60
                break
            elif secim=="3":
                oyuncu.bahsis_sans=50
                oyuncu.veresiye_sans=50
                break
            else:
                print("Lütfen tercihler içinden bir sayı giriniz.")
                continue
            
    
    dukkan_tercih()
            
main()

```
